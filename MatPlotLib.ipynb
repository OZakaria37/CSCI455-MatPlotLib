```python
dx = 0.01
dy = 0.02
x = np.arange(-2.0, 2.0, dx)
y = np.arange(-2.0, 2.0, dy)
X,Y = np.meshgrid(x,y)
def f(x,y):
  return (1- y**5 + x**5)*np.exp(-x**2-y**2)
plt.contour(X,Y,f(X,Y),8,colors='black')
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,f(X,Y),rstride=1,cstride=1)
```
